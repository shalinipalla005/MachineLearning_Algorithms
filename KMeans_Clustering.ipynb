```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=300, centers=4, random_state=42)
plt.scatter(X[:,0],  X[:, 1])
plt.title("Generated data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Now we will apply K-Means CLustering to this dataset

```python
k_means = KMeans(n_clusters= 4, random_state=42)

k_means.fit(X)

labels = k_means.labels_

centroids = k_means.cluster_centers_

plt.scatter(X[:,0], X[:,1], c=labels, cmap= 'viridis')
plt.scatter(centroids[:,0], centroids[:,1], s= 300, c="red", marker='X')
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Let's find optimal k value using Elbow Method

```python
inertia = []
```

```python
for k in range(1,10):
    k_means = KMeans(n_clusters=k, random_state=42)
    k_means.fit(X)
    inertia.append(k_means.inertia_)

plt.plot(range(1,10), inertia, marker = 'o')
plt.title("Elbow Method")
plt.xlabel("NUmber of clusters (k)")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()
```

From the curve above we can clearly see that the elbow point is at k = 3 --> the place where adding more clusters yields diminishing returns


So, the optimal number of clusters = 3
